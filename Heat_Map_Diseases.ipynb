```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pune = pd.read_csv('/content/drive/My Drive/dis.csv')
```

```python
pune.head()
```

```python
# @title Death/Infected

from matplotlib import pyplot as plt
import seaborn as sns
pune.groupby('Death/Infected').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
area_cases = pune.groupby('Area').sum()
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(area_cases, cmap='viridis', annot=True, fmt='g')
plt.title('Cases of Diseases in Pune Areas (2012-2019)')
plt.show()
```
